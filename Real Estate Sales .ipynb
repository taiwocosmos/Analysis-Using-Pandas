<h1>6. draw a conclusion for your new employer, the effect of time on the company sales?</h1>

```python
data['Profit'] = data['Sale Amount'] - data['Assessed Value'] 
```

```python
data
```

```python
data.groupby('List Year')['Profit'].sum().sort_values(ascending=False)
```

```python
correlation = data.corr(method='spearman')
correlation
```

<h6>With the table above, we have to cross relationship between the List Year and the 
    Sales Ratio to be Negative, which means that the time has no effect on the company sales</h> 

<h1>7. Kindly advise the company, which town should they stop selling Estate</h1>

```python
data.groupby('Town')['Profit'].sum().sort_values(ascending=True)
```

```python
sns.barplot(x='Town', y='Profit', data=data)
```

<h6>My conclusion is that the Real Estate Company should stop selling their Estates in 'Hamden' Town cause it has the lowest sales</h6>

<h1>9. We have different properties in our company, which property will you advise us to stop selling and which will you advise us to sell more of it?</h1>

```python
data.groupby('Property Type')['Profit'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Property Type', y='Profit', data=data)
```

<h6>Based on the Table above, we can see that the Property Type, 'Three Family' has a lower sales Ratio and brought 
the least profit to the company. Hereby, I advice to stop selling 'Three Family' property type and sell more of 'Residential' property type because it generates the highest Profit</h6>
