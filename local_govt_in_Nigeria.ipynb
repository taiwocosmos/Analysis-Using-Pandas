```python
import numpy as np
import pandas as pd
```

<h1>Reading Dataset for Local Governments in Nigeria</h1>

```python
LG = pd.read_csv('list_of_local_government_areas_of_nigeria.csv')
LG
```

```python
LG.isnull().any()
```

<h1>Cleaning of dataset LG</h1>

```python
LG.drop(columns='Unnamed: 3', inplace=True)
```

<h1>Reading Datasets for Nigeria poverty rate per local government</h1>

```python
povertyRate = pd.read_csv('Nigeria povertyrate per local gov.csv')
povertyRate
```

<h1>Cleaning of Dataset Povery rate</h1>

```python
povertyRate.drop(columns= 'Unnamed: 10', inplace=True)
```

<h1>Merging of the two Datasets</h1>

```python
newdata = pd.merge(LG, povertyRate, left_on='Local Government Area', right_on='NAME_23')
newdata
```

```python
newdata = newdata.set_index('Local Government Area')
newdata
```

```python
newdata.drop(columns='SNo', inplace=True)
```

```python
newdata
```

<h1>To check if there are any duplicate values</h1>

```python
newdata.duplicated()
```

```python
newdata.drop_duplicates(inplace=True)
```

<h1>To calculate how Population affects Poverty rate of each Local Government Area</h1>

```python
# We use the pandas correlation method to get it
correlation = newdata.corr(method='spearman')
correlation
```

```python
correlation.drop(index='ID_0', axis=0, inplace=True)
```

```python
correlation.drop(columns='ID_0', inplace=True)
```

```python
correlation
```

<h1>To calculate how population affects poverty rate, we need to find the correlation between both population and poverty
rate by employing the correlation method which consists of 'SPEARMAN, PEARSON AND KENDALL' after using the method and finding the
correlation we found out that population does not affect poverty because we had an inverse relationship(-0.055736), which is not between 0.5 and 1
</h1>

<h1>Sorting each local governments according to the order of their poverty rate</h1>

```python
newdata.sort_values('Poverty_Ra', ascending=False)
```

<h1>Percentage of Poverty</h1>

```python
newdata['Percentage(%)'] = newdata['Poverty_Ra']*100
newdata['Percentage(%)']
```

```python
newdata.sort_values('Percentage(%)', ascending=True)
```
